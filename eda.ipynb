```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set plot style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

print("‚úÖ Libraries imported")
```

```python
df = pd.read_csv('data/raw/breast_cancer.csv')

print(f"Dataset shape: {df.shape}")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
df.head()
```

```python
print("üìã Dataset Info:")
df.info()

print("\nüìä Statistical Summary:")
df.describe()
```

```python
# Target distribution
print("üéØ Target Distribution:")
print(df['target'].value_counts())
print("\nPercentage:")
print(df['target'].value_counts(normalize=True) * 100)

# Visualize
plt.figure(figsize=(6, 4))
df['target'].value_counts().plot(kind='bar', color=['salmon', 'lightblue'])
plt.title('Target Distribution (0=Malignant, 1=Benign)', fontsize=14)
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
print("‚ùì Missing Values:")
missing = df.isnull().sum()
if missing.sum() == 0:
    print("‚úÖ No missing values!")
else:
    print(missing[missing > 0])
    
# Visualize
plt.figure(figsize=(12, 4))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap (Yellow = Missing)')
plt.show()
```

```python
correlations = df.corr()['target'].sort_values(ascending=False)
print("üîó Top 10 Features Correlated with Target:")
print(correlations.head(11))  # 11 because target correlates with itself (1.0)

# Visualize top correlations
top_features = correlations.head(11).drop('target')  # Remove target itself
plt.figure(figsize=(8, 6))
top_features.plot(kind='barh', color='teal')
plt.title('Top 10 Features Most Correlated with Target', fontsize=14)
plt.xlabel('Correlation Coefficient')
plt.ylabel('Feature')
plt.axvline(x=0, color='black', linestyle='--', linewidth=0.8)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
selected_cols = df.columns[:11]  # First 10 features + target
corr_matrix = df[selected_cols].corr()

sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1)
plt.title('Feature Correlation Heatmap', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
top_4_features = correlations.head(5).drop('target').index[:4]

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()

for idx, feature in enumerate(top_4_features):
    # Split by target class
    df[df['target']==0][feature].hist(ax=axes[idx], alpha=0.6, 
                                       label='Malignant', bins=20, color='red')
    df[df['target']==1][feature].hist(ax=axes[idx], alpha=0.6, 
                                       label='Benign', bins=20, color='blue')
    axes[idx].set_title(f'{feature} Distribution by Class')
    axes[idx].set_xlabel(feature)
    axes[idx].set_ylabel('Frequency')
    axes[idx].legend()

plt.tight_layout()
plt.show()
```
